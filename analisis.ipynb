# Análisis con diagrama de dispersión con tiempos de ejecución de la máquina en función del tamaño de entrada

Tamaño de entrada de Fibonacci de 0 500

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

fibonacci_run_data = pd.read_csv('fibonacci_results.csv')

plt.figure(figsize=(16, 9))
plt.scatter(fibonacci_run_data['Input Decimal'], fibonacci_run_data['Tiempo Ejecucion (ms)'], color='blue', edgecolors='k', s=20, alpha=0.6)
# plt.yscale('log')
plt.xticks(np.arange(0, 501, step=25), rotation=45)
plt.title('Tiempo de ejecución vs Input decimal', fontsize=16)
plt.xlabel('Input (decimal)', fontsize=14)
plt.ylabel('Tiempo de ejecución (ms, log)', fontsize=14)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()

```
